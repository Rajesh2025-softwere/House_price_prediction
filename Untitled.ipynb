```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv("Bengaluru_house_Data.csv")
data
```

```python
data.drop(columns=['availability' , 'society' , 'area_type' , 'balcony'] , inplace=True)
data.head()
```

```python
data.shape
```

```python
data.dropna(inplace=True)
```

```python
data.isnull().sum()
```

```python
data.location.value_counts()
```

```python
data['location'] = data['location'].apply(lambda x: x.strip())
```

```python
data['location'].value_counts()
```

```python
location_state = data.groupby('location')['location'].agg('count')
```

```python
location_lessthen_10 = location_state[location_state <= 10]
```

```python
location_lessthen_10
```

```python
data['location'] = data['location'].apply(lambda x:'other' if x in location_lessthen_10 else x )
```

```python
data['location'].value_counts()
```

```python
# Convert the bhk in integer 
data['bedrooms'] = data['size'].str.split().str.get(0).astype(int)
```

```python
data.total_sqft.unique()
```

```python
def convertRange(x):
    try:
        temp = str(x).split('_')
        if len(temp) == 2:
            return (float(temp[0]) + float(temp[1])) / 2
        return float(x)
    except:
        return None
```

```python
data['total_sqft'] = data['total_sqft'].apply(convertRange)
```

```python
data.total_sqft.unique()
```

```python
data.dropna(inplace=True)
```

```python
data['sqft_per_bed'] = data['total_sqft']/data['bedrooms']
```

```python
data['sqft_per_bed'].describe()
```

```python
data2 = data[data['sqft_per_bed'] >= 300]
```

```python
data2['price_per_sqft'] = data2['price']*100000 / data2['total_sqft']
```

```python
data2
```

```python
data2['price_per_sqft'].describe()
```

```python
data3 = data2[data2['price_per_sqft'] >= 2000]
```

```python
data3
```

```python
data3.drop(columns=['size' , 'sqft_per_bed', 'price_per_sqft'] , axis=1 , inplace=True)
```

```python
data3
```

```python
from sklearn.preprocessing import OneHotEncoder , StandardScaler
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
from sklearn.pipeline import make_pipeline
from sklearn.compose import make_column_transformer
```

```python
 col_tran = make_column_transformer((OneHotEncoder(sparse_output=False),['location']), remainder='passthrough')
```

```python
lr = LinearRegression()
```

```python
scalar = StandardScaler()
```

```python
X = data3.drop(columns=['price'])
y = data3['price']
```

```python
X_train, X_test , y_train , y_test = train_test_split(X , y , test_size=0.2)
```

```python
model = make_pipeline(col_tran , scalar , lr)
```

```python
model.fit(X_train , y_train)
```

```python
model.score(X_test , y_test)
```

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
```

```python
tree = DecisionTreeRegressor()
ens = RandomForestRegressor(n_estimators=100)
```

```python
model2 = make_pipeline(col_tran , scalar , tree)
model2.fit(X_train, y_train)
model2.score(X_test, y_test)
```

```python
model3 = make_pipeline(col_tran , scalar , ens)
model3.fit(X_train, y_train)
model3.score(X_test, y_test)
```
